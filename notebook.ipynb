# Bike Sharing Demand
This dataset consists of the number of public bikes rented in Seoul's bike sharing system at each hour. It also includes information about the weather and the time, such as whether it was a public holiday.


```python
import pandas as pd
seoul = pd.read_csv("data/SeoulBikeData.csv")
seoul.head(100)
```

```python
seasons = ["Summer","Winter","Spring","Autumn"]
for season in seasons:
    seoul_season = seoul[seoul["Seasons"]==season]
    hour_season = seoul_season["Hour"]
    y_season = seoul_season["Rented Bike Count"]
    print(f'Showing different graphs and stats for {season}')
    plt.xlabel("Hours in a day")
    plt.ylabel("Total number of bikes rented")
    plt.scatter(hour_season,y_season)
    plt.show()
    average_bikes_count_season = seoul_season.groupby("Hour")["Rented Bike Count"].mean()
    print(average_bikes_count_season)
    sorted_average= average_bikes_count_season.sort_values(ascending=False)
    print("The bikes rented in descening order is as: ")
    print(sorted_average)
    print("The maximum bikes rented in a day, minimum and the means are: ")
    print(sorted_average.max(),sorted_average.min(),sorted_average.mean())
    plt.plot(average_bikes_count_season.index,average_bikes_count_season)
    plt.title(f'Total number of bikes rented during different times in a day in {season}')
    plt.xlabel("Times in a day")
    plt.ylabel("Total number of bikes rented")
    plt.show()
```

```python

X= seoul["Temperature(C)"].values
X = X.reshape(-1,1)
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X,y)
predictions = reg.predict(X)
plt.scatter(X,y)
plt.plot(X,predictions, color="red", label="Regression Line")
plt.title("Temperature vs Rented Bikes")
plt.xlabel("Temperature (C)")
plt.ylabel("Rented Bikes Count")
plt.legend()
plt.show()
```

```python
seasons = ["Summer","Winter","Spring","Autumn"]
for season in seasons:
    seoul_season = seoul[seoul["Seasons"]==season]
    X = seoul_season["Temperature(C)"].values
    X = X.reshape(-1,1)
    y_season = seoul_season["Rented Bike Count"]
    from sklearn.linear_model import LinearRegression
    reg = LinearRegression()
    reg.fit(X,y_season)
    predictions = reg.predict(X)
    plt.scatter(X,y_season)
    plt.plot(X,predictions, color="red", label="Regression Line")
    plt.title(f'Temperature vs Rented Bikes during {season}')
    plt.xlabel("Temperature (C)")
    plt.ylabel("Rented Bikes Count")
    plt.legend()
    plt.show()
    mean_bikes = y_season.mean()
    max_bikes = y_season.max()
    min_bikes = y_season.min()
    print("The mean, maximum and minimum rented bikes count are: ")
    print(mean_bikes)
    print(max_bikes)
    print(min_bikes)

```

```python
y = seoul["Rented Bike Count"]
from scipy.stats import pearsonr
cols = ['Humidity(%)', 'Wind speed (m/s)',
       'Visibility (10m)', 'Dew point temperature(C)',
       'Solar Radiation (MJ/m2)', 'Rainfall(mm)', 'Snowfall (cm)', 'Seasons',
       'Holiday', 'Functioning Day']

cols_score = []

for col in cols:
    x = seoul[col]
    if x.dtype == 'object':
        x = x.astype('category').cat.codes
    correlation = pearsonr(x,y)
    cols_score.append(correlation.statistic)
    
print(cols_score)

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.bar(cols,cols_score)
plt.title("Relation between different factors and the rented bike count")
plt.xlabel("Different factors")
plt.ylabel("Correlation index")
plt.xticks(rotation = 45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Bin the visibility into ranges
seoul['Visibility_Binned'] = pd.cut(seoul['Visibility (10m)'], bins=10)  # 10 equal-width bins

# Calculate the mean bikes rented for each visibility bin
binned_data = seoul.groupby('Visibility_Binned')['Rented Bike Count'].mean()

# Plot the binned data
binned_data.plot(kind='bar', color='skyblue', figsize=(10, 6))
plt.title("Average Rented Bikes by Binned Visibility")
plt.xlabel("Visibility (10m) Ranges")
plt.ylabel("Average Number of Bikes Rented")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
